```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/content/prime.csv')
```

Drop the unnecessary columns

```python
df.drop(['show_id'], axis = 1, inplace = True)
df.drop(['description'], axis = 1, inplace = True)
df
```

Missing Values

```python
#Checking if there is any missing values
df.isnull()
```

```python
#Checking total number of missing values in the dataset
df.isnull().sum()
```

```python
# Replacing the Null values
df['director'].fillna('Unobtainable', inplace=True)
df['cast'].fillna('Unobtainable', inplace=True)
df['country'].fillna('Unobtainable', inplace=True)
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df['date_added']= df['date_added'].ffill()
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

```python
df.duplicated()
```

```python
duplicates = df[df.duplicated(keep = 'first')]
print(duplicates)
```

```python
missing_values = df[df.isnull().any(axis=1)]
print(missing_values)
```

```python
df.dropna(axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['release_year'], bins=30, color='skyblue')
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(8, 5))
df['rating'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
```

```python
correlation_matrix = df.corr()
print("Correlation Matrix:\n", correlation_matrix)
```

```python
genres = df['listed_in'].str.split(', ').explode().value_counts()
plt.figure(figsize=(12, 6))
genres.plot(kind='bar', color='salmon')
plt.title('Distribution of Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=90)
```

```python
countries = df['country'].str.split(', ').explode().value_counts()
plt.figure(figsize=(12, 6))
countries.head(10).plot(kind='bar', color='lightblue')
plt.title('Top 10 Countries with Most Content')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
```
